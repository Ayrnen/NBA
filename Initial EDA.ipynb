## Run a multi-variable linear regression model using the points from the 4 quarters and the total points scored

```python
q1 = np.array(dfHome['Quarter 1'])
q2 = np.array(dfHome['Quarter 2'])
q3 = np.array(dfHome['Quarter 3'])
q4 = np.array(dfHome['Quarter 4'])

x = np.stack((q1, q2, q3, q4), axis=1)
y = np.array(dfHome['Total Points'])
model = LinearRegression().fit(x, y)

print(f"coefficient of determination: {model.score(x, y)}")
print(f"intercept: {model.intercept_}")
print(f"slope: {model.coef_}")
```

---
#### Findings:
* The collection of points from the 4 quarters is entirely responsible for the Total Points
* There is no direct correlation between quarters 
* From Heatmap / Pearson Coefficient we can see that each quarter has a ~54-50% chance of being correlated to the Total Points
    - This decreases with each progressive quarter
* From the Coefficient of Determination ($r^2$) we can see that (if correlated) each quarter is responsible for ~30-24% of the Total Points
    - This decreases with each progressive quarter
* With an intercept ranging from ~82-86 we can predict that in 3 quarters ~84 points will be scored
    - Intercept increases with each progressive quarter
* The slope consistently drops from Q1-Q4 despite the fact it hypothetically "should be" at 1.0.
    - All quarters' slopes are above 1.0
    - Despite this consistency, the slope of Q1 and Q2 being 1.086 and 1.059 respectively are significant enough to pay attention to.
* The data shows that points scored in Quarter 1 are over 5% more valuable than points scored in Quarter 4 with the only clear explanation being the 5% difference in mean between the two quarters.
    - Still ~4% more valuable than Q3 with a mean difference that is a fraction of a percent
    - All have same median


---
Despite Quarter 1 having consistently higher values for correlation and strength, the measures of center and distribution between the quarters is relatively even. This is weird because there is no clear sign showing that earlier quarters score more points, and therefore there is no reason different quarters should have different values. Some possible explanations for this are:
* The tone set by the earlier quarters is indicative of how the rest of the game will play out
* The data from blow-out games where the losing team gives up mid-game is affecting the data
