```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('TeamLocations.csv')
```

```python
sns.scatterplot(data = df, x = 'Longitude', y = 'Latitude', hue = 'Conference')
```

```python
from sklearn.cluster import KMeans
x = df['Longitude']
y = df['Latitude']

data = list(zip(x, y))
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(data)

plt.scatter(x, y, c=kmeans.labels_)
plt.show()
```
